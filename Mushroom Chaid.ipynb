Train - Test Splitting

```python
from sklearn.model_selection import train_test_split
X=mushroom.drop('class',1)
y=mushroom['class']
X_train, X_test,y_train, y_test = train_test_split(X,y ,random_state=104,test_size=0.4,shuffle=True)
```

Main Function for building the model

```python
def chisquare(x, column1, y):
    df=x
    df['target']=y
    target='target'
    ct={}
    for i in df[column1].unique():
        ct[i]=df[df[column1]==i][target].value_counts()
    ct=pd.DataFrame(ct).T
    ct=ct.fillna(0)
    ct['total']=ct[0]+ct[1]
    ct["expected"]=ct['total']/2
    ct["chisquare_0"]=np.sqrt(((ct[0]-ct['expected'])**2)/ct['expected'])
    ct["chisquare_1"]=np.sqrt(((ct[1]-ct['expected'])**2)/ct['expected'])
    return ct 

def best_feature(x,y, bf=''):
    df=x
    df['target']=y
    target='target'
    chi2={}
    listofcolumns=list(df.columns)
    listofcolumns.remove(target)
    if bf not in listofcolumns:
        listofcolumns=listofcolumns
    elif bf in listofcolumns:
        listofcolumns.remove(bf)
    for i in listofcolumns:
        a=pd.DataFrame(chisquare(x, i, y))
        chi2[i]=np.sum(a['chisquare_0']+a['chisquare_1'])  
    feature=pd.DataFrame(chi2,index=['chi2']).T
    return feature['chi2'].idxmax()

tree,route={},[]
def split(x, y, path=None):
    df=x
    df['target']=y
    target='target'
    if path is None:
        path = []
    bf = best_feature(x, y)
    for i in df[bf].unique():
        branch = f"{bf}->{i}"
        current_path = path + [branch]
        if df[df[bf] == i][target].nunique() == 1:
            decision = df[df[bf] == i][target].unique()[0]
            route.append((current_path, decision))
            print(f"{current_path} : {decision}")
        else:
            tree[branch] = df[df[bf] == i].drop(bf, axis=1)
#             print(f"{current_path} : {branch}")
            split(tree[branch], tree[branch][target], current_path)
    return route

def predict(x, route):
    df=x
    predictions = []
    for i, row in df.iterrows():
        for path, decision in route:
            a=0
            for branch in path:
                col, val = branch.split('->')
                if str(row[col]) != val:
                    a=1
                    break
            if a==0:
                predictions.append(decision)
                break
        else:
            predictions.append(max(set(predictions), key = predictions.count))
    return predictions
```

Decision Tree

```python
decision=split(X_train,y_train)
```

Taking the predictions

```python
y_pred=predict(X_train,decision)
```

```python
y_test_pred=predict(X_test,decision)
```

Computing the accuracy

```python
train_accuracy = np.mean(y_pred == y_train)
print(f"train Accuracy: {train_accuracy:.2%}")
```

```python
test_accuracy = np.mean(y_test_pred == y_test)
print(f"test Accuracy: {test_accuracy:.2%}")
```
